```python
import json
from collections import Counter

path = "./data/train/generated_qa_pairs.json"

with open(path) as f:
    data = json.load(f)

print("num_examples:", len(data))

for row in data[:10]:
    print(row)

ans = Counter(x["answer"] for x in data)

print("behind:", ans["behind"])
print("back:", ans["back"])
print("front:", ans["front"])
```

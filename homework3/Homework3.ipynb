```python
from homework.sft import load
llm = load()

test_q = "How many meters are in 2 kilometers?"

raw_output = llm.batched_generate([llm.format_prompt(test_q)], temperature=0.5)

print(f"PROMPT SENT: {llm.format_prompt(test_q)}")
print("RAW LIST LENGTH:", len(raw_output))
for i, r in enumerate(raw_output):
    print(f"\n--- RAW[{i}] ---\n{repr(r)}")

# pick the last one (often the “real completion” is last)
candidate = raw_output[-1]
print("\nCANDIDATE:", repr(candidate))
print("PARSED:", llm.parse_answer(candidate))
```

```python
import json

def filter_rft_data(input_path="data/rft.json", output_path="data/rft_clean.json"):
    with open(input_path, "r") as f:
        data = json.load(f)

    # Keep only samples that contain the required tag
    clean_data = [sample for sample in data if "<answer>" in sample[2]]

    with open(output_path, "w") as f:
        json.dump(clean_data, f, indent=2)

    print(f"Cleaned dataset: {len(clean_data)} samples remaining (Removed {len(data) - len(clean_data)})")

filter_rft_data()
```
